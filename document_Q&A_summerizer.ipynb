## Text Splitting

Split long text into smaller chunks to improve embedding quality and retrieval accuracy

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_spliter = RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap = 0,
    add_start_index=True,
)
```

```python
chunks = text_spliter.split_documents(document_data)
```

```python
print(type(chunks), len(chunks))
```

```python
#print(chunks[2].page_content, '\n', len(chunks[2].page_content))
#print(chunks[0].page_content)
for i in chunks:
  print(len(i.page_content),end=' ')
```

##(5) Vector store

## Generate Embeddings

Convert text chunks into numerical embeddings that capture semantic meaning.

```python
!pip install langchain-pinecone
```

```python
from langchain_pinecone import PineconeVectorStore
```

```python
pc_vectorstore = PineconeVectorStore.from_documents(
      chunks,
      index_name = "pcindx",
      embedding = pic_embedding
)
```

```python
hf_vectorstore = PineconeVectorStore.from_documents(
      chunks,
      index_name = "hfindx",
      embedding = hgf_embeddings
)
```

#(6) Retriever

```python
Q1="What are the essential components of a symmetric cipher, and how does it differ from asymmetric encryption?",
Q2="Explain the working of Caesar Cipher, Playfair Cipher, and Rail Fence Cipher with examples.",
Q3 ="What are the different types of security attacks, and how do active and passive attacks differ?"
```

```python
query = [Q1,Q2,Q3]
```

## Store Embeddings in Vector Store

Store embeddings in a vector database to enable similarity search.

```python
context = pc_vectorstore.similarity_search(str(query[0]),k=2)
```

```python
print(type(context), len(context))
```

```python
print(context[0])
```

```python
print(context[1])
```

```python
context_1 = pc_vectorstore.similarity_search(str(query[1]),k=2)
```

```python
print(context_1[1])
```

```python
print(context_2[1])
```

### Question Answering

Use an LLM to generate answers strictly based on retrieved document context.

#(7) Design Prompt

```python
from langchain import hub
rag_prmt = hub.pull("rlm/rag-prompt")
```

```python
print(rag_prmt.messages)
```

```python
msg = rag_prmt.invoke({"context": context, "question" : query}).to_messages()
```

```python
print(msg)
```

###  Summarization

Generate concise summaries of the document using the same retrieval-based grounding.
